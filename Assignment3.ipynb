```python
import os


def processfiles(directory):
    foler_name = os.listdir(directory)
    result = []
    for file in foler_name:
        path = directory + "/" + file
        with open (path, encoding = "ISO-8859-1") as f:
            text = f.read()
            result.append(text)
    return result

directory = "/Users/alexwang/Documents/NLP/Assignment3/Assignment Corpus"
corpus = processfiles(directory)


n = 0
for essay in corpus:
    n = n + len(essay) 
print(n)
    
```

```python
import pandas as pd 

data = {
    'Text': corpus
        }
turkey= pd.DataFrame(data)
turkey

```

```python
import re
from nltk.stem import *
from nltk.tokenize import word_tokenize
from gensim.parsing.preprocessing import remove_stopwords

def process_document(document):
    cleaned_document = re.sub(r'[^\w\s]','',document)
    cleaned_document = remove_stopwords(cleaned_document.lower())
    cleaned_document = word_tokenize(cleaned_document)
    wnl = WordNetLemmatizer()
    lemmas = []
    for w in cleaned_document:
        lemma = wnl.lemmatize(w)
        lemmas.append(lemma)    
    return lemmas

documents = turkey["Text"]

#Create a document vector
doc_vectors=[process_document(document)
             for document in documents]

#Print the document and the corresponding document vector to compare
print(documents[1])
print(doc_vectors[1])

```

```python
from gensim import corpora
dictionary = corpora.Dictionary(doc_vectors)

print("Dictionary created :")
dictionary.token2id
```

```python
bow_corpus = [dictionary.doc2bow(doc_vector) 
          for doc_vector in doc_vectors]

#Review the corpus generated
print(doc_vectors[1])
print(bow_corpus[1])
```

```python
import gensim

# µs = microsecond = 1/1000 of a second
TOTAL_TOPICS = 14  # Change this value to alter the number of topics the model includes
lsi_bow = gensim.models.LsiModel(bow_corpus, id2word=dictionary,
num_topics=TOTAL_TOPICS, onepass=True, chunksize=1740, power_iters=1000)
```

```python
for topic_id, topic in lsi_bow.print_topics(num_topics=14, num_words=20):
    print('Topic #'+str(topic_id+1)+':')
    print(topic)
    print()
```

```python
for n in range(TOTAL_TOPICS):
    print('Topic #'+str(n+1)+':')
    print('='*50)
    d1 = []
    d2 = []
    for term, wt in lsi_bow.show_topic(n, topn=9):
        if wt >= 0:
            d1.append((term, round(wt, 3)))
        else:
            d2.append((term, round(wt, 3)))
    print('Direction 1:', d1)
    print('-'*50)
    print('Direction 2:', d2)
    print('-'*50)
print()
```

```python
#Shape of the three major matrices (U, S, and VT)
term_topic = lsi_bow.projection.u
singular_values = lsi_bow.projection.s
topic_document = (gensim.matutils.corpus2dense(lsi_bow[bow_corpus],
len(singular_values)).T / singular_values).T
term_topic.shape, singular_values.shape, topic_document.shape
```

```python
import pandas as pd
import numpy as np
#document-topic matrix to help us see the proportion of each topic per document (a larger proportion means the topic is more dominant in the document).

document_topics = pd.DataFrame(np.round(topic_document.T, 3),
                               columns=['Topic '+str(i) for i in range(1, TOTAL_TOPICS+1)])
document_topics.head(15)
```

```python
#View documents 1-3 with topic allocation. 
document_numbers = 101 # Change this list to match the number of documents
for document_number in range(document_numbers):
    top_topics = list(document_topics.columns[np.argsort(-np.absolute(document_topics.iloc[document_number].values))[:3]])
    print('\nDocument #'+str(document_number)+':')
    print('Dominant Topics (top 3):', top_topics)
    print('Document Text:')
    print(documents[document_number][:500])
print()
```
